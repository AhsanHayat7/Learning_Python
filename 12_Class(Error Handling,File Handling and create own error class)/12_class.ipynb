```python
class StudentCard():
    def __init__(self, roll_no : int, name : str, age : int ) -> None:
        if age < 18 or age > 60:
            raise StudentAgeError("You are not eligible for this program!")
        self.roll_no = roll_no
        self.name = name
        self.age = age


class StudentAgeError(Exception): # custom error
    pass

student1 = StudentCard(4,"Watson",23)

print(student1.name,student1.roll_no,student1.age)
```

```python
with open("./abc.txt") as file: # type TextIO
    print(type(file))
    print(file.read()) # read lines
print("\nHappy coding!")    
```

```python
with open("./abc.txt") as file: # type TextIO
    print(type(file))
    print(file.readline(), end="") # read only one line  at a time(first line)
print("\nHappy coding!")  
```

```python
with open("./abc.txt") as file: # type TextIO
    print(type(file))
    print(file.readlines()) # read all lines
print("\nHappy coding!")  
```

```python
with open("./abc.txt") as file: # type TextIO
    print(type(file))
    print(file.readlines()[:1]) # read lines upto 1 
print("\nHappy coding!")  
```

```python
with open("./abc.txt", 'r') as file: # type TextIO 
    print(type(file))
    print(file.readlines()[:1]) # read lines upto 1 
print("\nHappy coding!")  
```

```python
with open("./abc.txt", 'r') as file: # type TextIO 
    print(type(file))
    file.write("Hello Coder")
```

```python
with open("./abc.txt", 'w') as file: # type TextIO 
    print(file.read())
    file.write("Hello Coder")
```
